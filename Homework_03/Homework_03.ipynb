## Задание 1

```python
import pandas as pd

books = pd.DataFrame({
    'Название': ['Мастер и Маргарита', '1984', 'Преступление и наказание', 'Sapiens', 'Atomic Habits', 'Дюна', 'Тонкое искусство пофигизма'],
    'Автор': ['Булгаков', 'Оруэлл', 'Достоевский', 'Харари', 'Клир', 'Герберт', 'Мэнсон'],
    'Год издания': [1967, 1949, 1866, 2011, 2018, 1965, 2016],
    'Цена': [850, 720, 650, 1200, 1500, 980, 1100],
})
print(books)
print('Средняя цена:', books['Цена'].mean())
print(books[books['Год издания'] > 2015])
print(books.sort_values('Цена'))
```

## Задание 2

```python
orders = pd.read_csv('data/orders.csv')
print(orders.head(10))
print(orders['Клиент'].value_counts())
print('Максимальная сумма:', orders['Сумма'].max())
print('Минимальная сумма:', orders['Сумма'].min())
print('Общая сумма заказов:', orders['Сумма'].sum())
```

## Задание 3

```python
food = pd.DataFrame({
    'Продукт': ['Яблоко', 'Банан', 'Куриная грудка', 'Говядина', 'Лосось', 'Рис', 'Гречка', 'Творог', 'Сыр', 'Миндаль', 'Овсянка'],
    'Категория': ['Фрукты', 'Фрукты', 'Мясо', 'Мясо', 'Рыба', 'Крупы', 'Крупы', 'Молочное', 'Молочное', 'Орехи', 'Крупы'],
    'Калории': [52, 89, 165, 250, 208, 130, 343, 98, 402, 579, 389],
    'Белки': [0.3, 1.1, 31, 26, 20, 2.7, 13, 18, 25, 21, 17],
})
print(food)
print(food[food['Калории'] > 300])
print(food.groupby('Категория')['Белки'].mean())
print(food.sort_values('Калории', ascending=False))
```

## Задание 4

```python
work = pd.DataFrame({
    'Имя': ['Анна', 'Борис', 'Анна', 'Виктор', 'Борис', 'Анна', 'Виктор', 'Дмитрий'],
    'Проект': ['Альфа', 'Альфа', 'Бета', 'Гамма', 'Бета', 'Гамма', 'Альфа', 'Бета'],
    'Часы': [12, 8, 15, 20, 10, 7, 5, 18],
})
print(work)
print(work.groupby('Имя')['Часы'].sum())
print(work.groupby('Проект')['Часы'].sum())
print('Больше всего часов у:', work.groupby('Имя')['Часы'].sum().idxmax())
```
