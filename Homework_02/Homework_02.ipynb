### Отчёт по продажам

*В этом разделе мы считаем итоги по простым данным.*

```python
sales = [120, 135, 100, 150, 90, 200, 170]
sales = np.array(sales)
print("Общая сумма:", sales.sum())
print("Среднее:", round(sales.mean(), 2))
print("Максимум:", sales.max())
print("Минимум:", sales.min())
```

|------------|----------|
